```python
import pandas as pd
import numpy as np
import json


with open('d1_user_dict.json','r',encoding="UTF-8") as f:
    user_dict = json.load(f)
with open('d1_item_dict.json','r',encoding="UTF-8") as f:
    item_dict = json.load(f)

user_dict = {int(v): k for k, v in user_dict.items()}    
item_dict = {v: k for k, v in item_dict.items()} 

file = '0_matrix.npy'
temp = np.load(file)
df = pd.DataFrame(temp)
```

```python
def item_change(x):
    return x.map(lambda x:item_dict[x])


def parsing_result(df):
    df = df.copy()
    
    df_n = df.apply(item_change)
    df_n.rename(index=user_dict, inplace =True)
    return df_n
    
def parsing_matrix(df):
    df = df.copy()
    
    df.rename(index=user_dict, inplace =True)
    df.rename(columns=item_dict, inplace =True)
    return df
    

df_matrix = parsing_matrix(df)
df_matrix
```
